```python
df_corr = data_2022.select_dtypes(include=['int64','float64'])

correlation_matrix = df_corr.corr()
# Establecer un umbral para considerar una correlación como fuerte
plt.figure(figsize=(20,6))
sns.heatmap(correlation_matrix,annot=True,cmap='coolwarm',vmin=-1,vmax=1)
plt.title('Mapa de calor de correlacion')
plt.show()
threshold = 0.6

# Encontrar pares de columnas con correlación fuerte
strong_correlations = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            colname1 = correlation_matrix.columns[i]
            colname2 = correlation_matrix.columns[j]
            correlation_value = correlation_matrix.iloc[i, j]
            strong_correlations.append((colname1, colname2, correlation_value))

# Imprimir los resultados
print("Pares de columnas con correlación fuerte:")
for col1, col2, corr in strong_correlations:
    print(f"{col1} y {col2}: {corr:.2f}")
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
from shapely.geometry import Point, LineString
```

```python
data_gps = data_2022[(data_2022['GPS_LONGITUD'] != 0) & (data_2022['GPS_LATITUD'] != 0)][['ID_CUENTA', 'GPS_LONGITUD', 'GPS_LATITUD']] 
data_gps
```

```python
# Convertir los datos a un GeoDataFrame
gdf = gpd.GeoDataFrame(
    data_gps, 
    geometry=gpd.points_from_xy(data_gps.GPS_LONGITUD, data_gps.GPS_LATITUD)
)

# Crear una figura y un eje para el gráfico
fig, ax = plt.subplots(figsize=(10, 10))

# Graficar los puntos de GPS
gdf.plot(ax=ax, color='blue', markersize=50, label='Puntos GPS')

# Opcional: Conectar los puntos con líneas para visualizar una ruta
if len(gdf) > 1:
    line = LineString(gdf.geometry)
    gpd.GeoSeries([line]).plot(ax=ax, color='red', linestyle='--', linewidth=2, label='Ruta')

# Configurar el mapa
ax.set_title('Visualización de Puntos de GPS y Ruta')
ax.set_xlabel('Longitud')
ax.set_ylabel('Latitud')
ax.legend()
ax.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
from folium.plugins import HeatMap
```

```python
data_gps_por_mes = data_2022[(data_2022['GPS_LONGITUD'] != 0) & (data_2022['GPS_LATITUD'] != 0)][['ID_CUENTA', 'GPS_LONGITUD', 'GPS_LATITUD','FECHA_INICIO_FORMAT']] 
data_gps_por_mes
```

```python
#TODO EL MAPA QUE RECOMIENDO PARA PONER EN EL CASO DE PONER UN MAPA EN PARTICULAR ES mapa_calor_2022-04 que se encuentra en reports/figures/1.1 - act_rep_initial-data/Mapa_calor_meses 
# Extraer el mes y el año
data_gps_por_mes['Year_Month'] = data_gps_por_mes['FECHA_INICIO_FORMAT'].dt.to_period('M')

# Iterar sobre cada mes
for period in data_gps_por_mes['Year_Month'].unique():
    # Filtrar los datos para el mes actual
    monthly_data = data_gps_por_mes[data_gps_por_mes['Year_Month'] == period]

    # Verificar si el DataFrame filtrado tiene datos
    if not monthly_data.empty:
        # Crear un mapa centrado en la media de las coordenadas del mes
        m = folium.Map(location=[monthly_data['GPS_LATITUD'].mean(), monthly_data['GPS_LONGITUD'].mean()], zoom_start=13)

        # Convertir los datos en una lista de coordenadas
        heat_data = [[row['GPS_LATITUD'], row['GPS_LONGITUD']] for _, row in monthly_data.iterrows()]

        # Añadir el mapa de calor
        HeatMap(heat_data).add_to(m)

        # Guardar el mapa en un archivo HTML
        month_str = period.strftime('%Y-%m')
        filename = f'mapa_calor_{month_str}.html'
        m.save(filename)

        print(f"Mapa de calor para {month_str} guardado como '{filename}'")
    else:
        print(f"No hay datos disponibles para {period}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

# Supongamos que ya tienes tu DataFrame 'data_2022' agrupado por 'Year_Month'
# y con las columnas 'DIFERENCIA' y 'MONTO_COBRADO' en formato numérico.

# Lista de meses únicos
meses_unicos = data_2022_cobranza['Year_Month'].unique()

# Determinar el tamaño de la cuadrícula
num_meses = len(meses_unicos)
cols = math.ceil(math.sqrt(num_meses))  # Número de columnas (redondeado hacia arriba)
rows = math.ceil(num_meses / cols)      # Número de filas

# Configurar el tamaño de la figura
fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(4*cols, 5*rows))

# Asegurarse de que axes sea un array (en caso de que haya un solo gráfico)
axes = axes.flatten() if num_meses > 1 else [axes]

# Generar un gráfico de torta para cada mes
for i, mes in enumerate(meses_unicos):
    # Filtrar los datos para el mes actual
    data_mes = data_2022_cobranza[data_2022_cobranza['Year_Month'] == mes]
    
    # Valores para el gráfico
    valores = [data_mes['DIFERENCIA'].values[0], data_mes['MONTO_COBRADO'].values[0]]
    etiquetas = ['MONTO NO COBRADO', 'MONTO COBRADO']
    
    # Crear gráfico de torta en el subplot correspondiente
    axes[i].pie(valores, labels=etiquetas, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(mes)
    
    # Agregar los montos exactos debajo del gráfico
    monto_no_cobrado = f"MONTO NO COBRADO: ${valores[0]:,.2f}"
    monto_cobrado = f"MONTO COBRADO: ${valores[1]:,.2f}"
    
    # Añadir anotaciones
    axes[i].annotate(monto_no_cobrado, xy=(0, -1.2), xytext=(0, -1.2),
                     textcoords='axes fraction', ha='center', fontsize=10)
    axes[i].annotate(monto_cobrado, xy=(0, -0.9), xytext=(0, -0.9),
                     textcoords='axes fraction', ha='center', fontsize=10)

# Eliminar gráficos sobrantes si hay más subplots que meses
for j in range(i+1, len(axes)):
    fig.delaxes(axes[j])

# Ajustar el layout para evitar solapamientos
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
